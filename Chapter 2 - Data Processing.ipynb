```python
cat_data = array(['male', 'female', 'male', 'male', 'female', 'male', 'female', 'female'])
```

```python
def cat_to_num(data):
    categories = unique(data)
    features = []
    for cat in categories:
        binary = (data == cat)
        features.append(binary.astype("int"))
    return features
```
